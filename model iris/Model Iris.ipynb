```python
import numpy as np
import pandas as pd
import sklearn as sk

from sklearn import datasets
```

```python
# importando dos dados

data_set = datasets.load_iris()
X = pd.DataFrame(data= data_set.data, columns=data_set.feature_names)
Y = pd.DataFrame(data= data_set.target)

display(X)
display(Y)
print(data_set.target_names)
```

```python
# pré processamento

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = pd.DataFrame(data=scaler.fit_transform(X), columns= list(X.columns))
display(X)
```

```python
# modelando

from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier

X_train, X_test, y_train, y_test = train_test_split(X.values, Y.values.ravel(), random_state=1, 
                                                    test_size=0.8)

model = GradientBoostingClassifier()
model.fit(X.values,Y.values.ravel())
y_pred = model.predict(X_test)
model.score(X_test, y_test)
```

```python
# Avaliação

from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

print("----- matriz da confusão ----- \n")
display(confusion_matrix(y_test, y_pred))
print("\n\n\n")
print("----- relatório de resultados ----- \n")
print(classification_report(y_test, y_pred, target_names=data_set.target_names))

# resultados muito bons, não necessário fazer a tunagem.
```

```python
# salvando

import pickle

pickle.dump(model, open("modelos/iris_model.sav", "wb"))
pickle.dump(scaler, open("modelos/iris_scaler.sav", "wb"))
```
